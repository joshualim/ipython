## Previously, I addressed a misconception that effective heating elements should have a low resistance. Now, we extend this to think about how low should the resistance be?
Heat is generated in an electric circuit when current flows: the greater the current, the greater the heat output. For designing a heating element this tells us that we should pick something with a low resistance to allow lots of current to flow. The question is how low can we make the resistance of the element and have a good heating element: 10 Ohm, 1 Ohm, 0.001 Ohm?  

#### Back to the last problem
Recall the simple heating element circuit from the previous example, which is made of a resistor (with resistance, R), battery (supplying a voltage Vsupply) and connecting wires. 

<center> <img src="files/circuit1.png" /> </center>

In solving this problem previously we ignored two things:
1. the wires have some small resistance (Rwire).
2. the battery has an internal resistance.
Why do I care about these now? Well, as we make the resistance of the element closer to zero, we are effectively ignoring this so we should take care of these other contributions.

#### Why was ignoring the wire resistance and internal resistance ok before?
To answer this we have to revisit our simple circuit and start to think about the things we left out before. Our new circuit diagram looks is now shown where, we now also have Rwire and Rint (for the wire and internal resistances).
<center> <img src="files/circuit3.png" /> </center>
1. Previously we ignoring the resistance of the wire as the resistance of the element was much bigger and a small wire resistance made now difference to the current. We can see this by rearranging Ohm's law for a series circuit (and including Rwire): 

$I=\frac{V}{Rwire+Relement}$

for the case where Relement >> Rwire, we will have :

$I=\frac{V}{Relement}$

(if you don't believe me use Rwire=0.02 and Relement=10, with a 1V supply voltage)



2. 



Often, when solving a problem we will make some simplifying assumptions - it's important to be aware of these as we will see when we change things.

```python
## Calculate the resistance of a typical wire, Rwire
Cu_resistivity=1.68*10**-8 ## in Ohm meters
L=1                        ## in m
diameter = 0.5               ## in mm
area = (0.001*diameter/2)**2 * pi    ## in m2
Rwire = Cu_resistivity * L / area ## in Ohms
print "Rwire"
print Rwire
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import *

# set up problem
Vsupply=1  # set the supply voltage
Relementmin=0.001; Relementmax=1; Relementincr=(Relementmax-Relementmin)/1000  ## let the element resistance to change from 0 to 1000 Ohms
Relement=arange(Relementmin,Relementmax,Relementincr)

# define a function that calculates the power for elements of different resistances
def powercalc(Vsupply,Rwire,Relement):
    
    I=Vsupply/(Rwire+Relement)  # calculate current through circuit using Ohms law
    PowerElement=I**2*Relement  # calc power dissipated in heating element
    PowerWire=I**2*Rwire        # calc power dissipated in the wires
    PowerTot=PowerElement+PowerWire
        
    return PowerElement, PowerWire, PowerTot

PowerElement, PowerWire, PowerTot = powercalc(Vsupply,Rwire,Relement)
    
```

```python
# make a graph of results
plt.plot(Relement,PowerElement,label='Heating Element Power')
plt.plot(Relement,PowerWire,label='Power dissipated in wire')
plt.plot(Relement,PowerTot,label='Total Power')

plt.xlabel('Heating element resistance (Ohms)')
plt.ylabel('Power (W)')
plt.title(( 'V$_{supply}$ = '+str(Vsupply)+' V,  R$_{wire}$ = '+str(Rwire)+' Ohms' ))
plt.legend()
plt.ylim([0,10])  
plt.show()
```

```python
## calculate efficiency
efficiency= PowerElement/PowerTot  

plt.plot(Relement,efficiency)
plt.xlabel('Heating element resistance (Ohms)')
plt.ylabel('Efficency ')
plt.title(( 'Heating Element Efficiency' ))
plt.ylim([0,1])  
plt.show()
```

```python
## define a figure of merit
FOM = efficiency * PowerElement

plt.plot(Relement,FOM)
plt.xlabel('Heating element resistance (Ohms)')
plt.ylabel('FOM ')
plt.title(( 'Heating figure of merit based on power output and efficiency' ))
#plt.ylim([0,2])  
plt.legend()
plt.show()
```
